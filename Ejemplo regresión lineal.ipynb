## Formula para minimizar el error cuadrático medio (mínimos cuadrados ordinarios)

 $media((yr - ye)^2) $  # el error de todos los puntos en la gráfica
 
 $ \beta = (X^{T}X)^{-1}X^{T}Y$ # error representado mediante matrices


```
# Para graficar se conviertes los datos en matrices
X = np.array(boston.data[:, 5])
Y = np.array(boston.target)

# Se visualizan los puntos:
# alpha: trasparencia en los puntos
plt.scatter(X, Y, alpha=0.5)

# Se añaden columnas de uno para terminos independientes
# X traspuesta porque se desea tener 506 filas y 2 columnas
X = np.array([np.ones(506), X]).T

# Se hace la formula para calcular el error cuadratico medio
B = np.linalg.inv(X.T @ X) @ X.T @ Y # el @ se usa en python para multiplicacion matricial, el .T representa la traspuesta

# Graficar la linea de regresión entre x= [4,9]
plt.plot([4,9], [B[0]+ B[1] * 4, B[0]+ B[1]* 9 ], c="red")

# Mostrar la grafica
plt.show()
```
